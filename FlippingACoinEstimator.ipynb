```python
from sklearn.base import BaseEstimator, ClassifierMixin, RegressorMixin #Ridge,Lasso = RegressorMixin
import numpy as np

class FlippingACoinEstimator(BaseEstimator, ClassifierMixin): # using Classifier method


    def __init__(self):
        self.prob=0
        
    def fit(self, X, y):
        self.prob=float(sum(y))/len(y)

    def predict(self,X):
        our_results = np.random.rand(len(X),1)
        for i in range(len(our_results)):
            if our_results[i]<self.prob:
                our_results[i]=1
            else:
                our_results[i]=0
        return our_results
    

m=95
n=90
#Information about the company
inptL=np.random.rand(m,n) #  defines a matix of dimension m*n with decimal values:[0,1]
#Stock Collapsed
outptL=np.random.randint(2, size=(m, 1)) # defines a matix of dimension m*1 with integer values:{0,1}


inptT=np.random.rand(m,n)
outptT=np.random.randint(2, size=(m, 1))
ourCoin=FlippingACoinEstimator()

ourCoin.fit(inptL,outptL) # creates the model using training data, "L"

print ourCoin.predict(inptT) # predicting with real input from test data
print ourCoin.score(inptT,outptT) # comparing the output from the model with the real.. 
                                   # ...output from test data

    
    
print ourCoin.score(ourCoin.predict(inptT),outptT)
print ourCoin.prob


```
