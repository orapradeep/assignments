# Best fit line

The best fit line is a technique for using some known data to approximate some unknown data. We will use the known data to find a mathematical equation for a line of the form:

$$ y = mx + b $$

where $m$ is the **slope** and $b$ is the **intercept**. Once we construct this line, we can find approximations for points which we were not included in the original dataset.

Suppose we have some data $x$ and $y$:

```python
# This data structure is called a "numpy array"
# Is is much more efficient than python lists for numerical analysis
# It also offers a great number of methods for operating on numerical data
x = np.array([10., 10.5, 11.,  11.5, 12.])
y = np.array([3.0, 3.04, 3.08, 3.11, 3.2])
```

Let's plot our data to take a look.

```python
# Plot the data using matplotlib.pyplot
plt.plot(x,y,'s')    # 's' gives us square markers

# It's always nice to include labels! Visualization IS communication.
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data')

# Adjust the size of the viewing window
plt.xlim([9.5,12.5])
plt.ylim([2.9,3.3])
```

The process by which we determine the line of best fit is called **linear regression**. In future classes, we will extend the concept of linear regression to fit more data with patterns more complicated than a simple line. This is just the starting point!

A function to perform linear regression is provided in the package `scipy.stats`. 

```python
# Call linregress()
slope, intercept, r_value, p_value, std_err = linregress(x,y)

# The outputs r_value, p_value, and std_err are provided to help us
# decide if the quality of our best fit line is good.
# For now, let's focus on slope and intercept. More on these others later.
```

We now have the slope and intercept of our fit (i.e. $m$ and $b$ in $y=mx+b$). Let's plot the line on top of our data and have a look:

```python
# Choose some evenly spaced points in the range of interest
xfit = np.linspace(10,12,100)
# Evaluate our line of best fit (y=mx+b)
yfit = slope*xfit + intercept
```

```python
plt.plot(x,y,'s')
plt.plot(xfit,yfit,'k--')    # 'k--' gives us a black dashed line
plt.legend(["Data","Best fit"])

plt.xlim([9.5,12.5])
plt.ylim([2.9,3.3])
```

Suppose we wish to predict the value at $x=10.75$. We can inspect the chart above and approximate it. We can also evaluate this poing directly using the equation of a line.

```python
xpredict = 10.75
ypredict = slope*10.75 + intercept
print ypredict
```
